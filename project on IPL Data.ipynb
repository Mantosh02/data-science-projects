```python
battingFirst= data_ipl[data_ipl["WonBy"]=="Runs"]
battingFirst
```

```python
battingFirst.head()
```

```python
plt.figure(figsize=(5,5))
plt.hist(battingFirst["WonBy"]=="Runs")
plt.title("distribution of runs")
plt.xlabel("runs")
plt.show()
```

```python
#finding frequency of batting first team wins
battingFirst["WinningTeam"].value_counts()
```

```python
#making a bar plot for the top 3 teams with mosst wins after batting first
plt.figure(figsize=(7,7))
plt.bar(list(battingFirst["WinningTeam"].value_counts()[0:3].keys()),list(battingFirst["WinningTeam"].value_counts()[0:3]),color=["blue","yellow","purple"])
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.pie(list(battingFirst["WinningTeam"].value_counts()),labels=list(battingFirst["WinningTeam"].value_counts().keys()),autopct='%0.2f%%')
plt.show()
```
